```python
import matplotlib.pyplot as plt

x=[1,10,14,23,24,50,48,7,5,58,66,95,42,15,12,34,36,27,39,55]
range=(0,95)
y=15
plt.hist(x,y,range,color='blue',histtype='bar',rwidth=0.5)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('histogram plotting')
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[13,14,15]
y=[74.5,74.45,74.4]
plt.scatter(x,y,label='weight',color='green',marker='.',s=100)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title("weight day plotting")
plt.legend()
plt.show()
```

```python
x=['india','america','japan','china','russia']
y=[4,3,8,15,12]
co=['red','green','m','b','y']
plt.pie(y,labels=x,colors=co,startangle=90,shadow=True,explode=(0.2,0,0,0,0),autopct='%1.2f%%')
plt.legend()
plt.title('corruption')
plt.show()
```
