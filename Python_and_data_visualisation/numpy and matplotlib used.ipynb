```python
import numpy as np
import matplotlib.pyplot as plt

t=np.arange(0.0,2.1,0.1)

s= 1 + np.cos(2*np.pi*t)

plt.grid()
plt.plot(t,s,'--,g')
plt.xlabel("time(S)")
plt.ylabel("amplitude")
plt.title("cosine graph")

plt.show()
```

```python
x1=np.linspace(0.0,5.0)
x2=np.linspace(0.0,2.0)

y1=np.cos(2*np.pi*x1)*np.exp(-x1)
y2=np.cos(2*np.pi*x2)

plt.subplot(2,1,1)
plt.grid()
plt.plot(x1,y1,'o-')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("subplot1")
plt.show()

plt.subplot(2,1,2)
plt.grid()
plt.plot(x2,y2,'-')
plt.xlabel("time")
plt.ylabel("amplitude")
plt.title("subplot2")
plt.show()
```
